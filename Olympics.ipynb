### In every olympics, which US athlete has won the most total number of medals? Include the athlete's discipline?

```python
USA = oo[oo.NOC == 'USA']
```

```python
rank_athlete_us = USA.groupby(['Athlete', 'Discipline', 'Medal']).size().unstack('Medal', fill_value=0)
```

```python
rank_athlete_us['total'] = rank_athlete_us['Gold']+rank_athlete_us['Silver']+rank_athlete_us['Bronze']
```

```python
rank_athlete_us.sort_values(['Gold','Silver','Bronze','total'], ascending= False)[['Gold','Silver','Bronze','total']]
```

```python
plt.figure(figsize = (300,60))
rank_athlete_us.sort_values(['Gold','Silver','Bronze'], ascending= False)[['Gold','Silver','Bronze']].head(10).plot(kind= 'bar')
```

```python
rank_athlete_us.reset_index(inplace =  True)
```

```python
top_USA = [group.sort_values('total', ascending = False) for year,group in rank_athlete_us.groupby('Edition')]
top_USA
```
